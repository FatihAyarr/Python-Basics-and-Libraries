# Relational Plots with Matplotlib

* Scatter plot: shows the distribution of the relationship between two variables with data points.
* Lineplot: Shows the relationship between two variables continuously. Data points are connected to each other by lines (used in time series).

* s parameter: marker size
* c parameter: marker color, which variable it holds can also be written
* alpha: marker opacity

```python
plt.scatter(df.Age, df.Insulin, c=df.Overweight, s=40,
            alpha=1)
plt.colorbar(); # a scale showing which color corresponds to which value
plt.xlabel("Age")
plt.ylabel("Insulin")
plt.title("Relationship between Age and Insulin")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df.Age, df.Insulin, c=df.Overweight)
ax.set_xlabel("Age")
ax.set_ylabel("Insulin")
ax.set_title("Relationship between Age and Insulin")
plt.show()
```

# Histogram: Helps us interpret the distribution of numerical or categorical data.

```python
fig, ax = plt.subplots()
ax.hist(df.Age, label="Age", bins=5)
ax.set_xlabel("Age")
ax.set_ylabel("Number of Observations")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df.SkinThickness, label="SkinThickness", bins=5)
ax.set_xlabel("SkinThickness")
ax.set_ylabel("Number of Observations")
plt.show()
```

# Bar plot: Allows us to look at the characteristics of categorical data.

```python
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']
totalDeaths = [112596, 37312, 5971, 27136, 40597, 7449]
plt.bar(countries, totalDeaths)
plt.show()
```

```python
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']
totalDeaths = [112596, 37312, 5971, 27136, 40597, 7449]

plt.bar(countries, totalDeaths, width= 0.8, align='center',color='blue', edgecolor = 'red')

plt.legend(labels = ['Total Deaths'])
plt.title("Bar plot representing the total deaths by top 6 countries due to coronavirus")
plt.xlabel('Countries')
plt.ylabel('Deaths')
plt.savefig('1BarPlot.png')
plt.show()
```

```python
plt.figure(figsize=[7, 5])

plt.barh(['USA', 'Brazil', 'Russia', 'Spain', 'UK'], [2026493, 710887, 476658, 288797, 287399], label = "Danger zone", color = 'r')
plt.barh(['India', 'Italy', 'Peru', 'Germany', 'Iran'], [265928, 235278, 199696, 186205, 173832], label = "Not safe zone", color = 'g')

plt.legend()

plt.xlabel('Total cases')
plt.ylabel('Countries')
plt.title('Top ten countries most affected by\n coronavirus')
plt.savefig('2BarPlot.png')
plt.show()
```
